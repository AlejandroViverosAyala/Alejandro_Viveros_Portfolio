```python
import pyreadr

safety_data = pyreadr.read_r('safety.rda')

df = safety_data['All']
df.head()
```

|num_effects| The number of adverse effects experienced by a single individual

```python
import pandas as pd

order = ['sex', 'age', 'week', 'trx', 'wbc', 'rbc','adverse_effects','num_effects']
df['adverse_effects'] = df[['headache','ab.pain','dyspepsia','upper.resp.infect','coad']].any(axis=1)
df['num_effects'] = df[['headache','ab.pain','dyspepsia','upper.resp.infect','coad']].sum(axis=1)
df2 = df[order]
df2.head()
```

# The proportion of adverse effects differs significantly between the Drug and Placebo groups?

To answer this question, I'm going to compare the proportion of participants with adverse effects in two groups:

- Drug
- Placebo

And make two different hypothesis:

- **Null Hypothesis:** the proportion of adverse effects is equal in both groups
- **Alternative Hypothesis:** the proportion of adverse effects differs between each group

For this I'll start creating an contingency table:

```python
contingency_table = pd.crosstab(df['trx'],df['adverse_effects'])
contingency_table
```

```python
from scipy.stats import chi2_contingency

chi2_stat, p_value, dof, expected = chi2_contingency(contingency_table)

print("Chi-Square Statistic (χ²):", chi2_stat)
print("p-value:", p_value)
print("Degrees of Freedom:", dof)
print("Expected Frequencies:\n", pd.DataFrame(expected, columns=['No_Adverse_Effect', 'Adverse_Effect'], index=['Drug', 'Placebo']))
```

```python
import matplotlib.pyplot as plt
import numpy as np

observed = contingency_table.values

expected = np.array(expected)

x_labels = ['Drug - No Adverse Effect', 'Drug - Adverse Effect', 'Placebo - No Adverse Effect', 'Placebo - Adverse Effect']
x = np.arange(len(x_labels))

observed_flat = observed.flatten()
expected_flat = expected.flatten()

width = 0.4

plt.bar(x - width/2, observed_flat, width, label='Observed', color='skyblue')
plt.bar(x + width/2, expected_flat, width, label='Expected', color='orange')

plt.xticks(x, x_labels, rotation=45, ha='right')
plt.ylabel('Frequency')
plt.title('Observed vs. Expected Frequencies')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
print("P-value:", p_value)
```

This result suggests that the Drug and Placebo are likely to have a similar rate of adverse effects, and the difference observed in the data could just be due to random chance rather than an actual effect of the treatment.

Since the p-value was 0.9549 (much greater than the typical significance level of 0.05), it indicates that the number of adverse effects is independent of the treatment group. In other words, there is no significant relationship between whether a patient received the drug or the placebo and whether they experienced adverse effects.

```python
from scipy.stats import ttest_ind, shapiro, levene

drug_group = df[df['trx'] == 'Drug']['age']
placebo_group = df[df['trx'] == 'Placebo']['age']

print('Shapiro-Wilk Test for Normality: ')
print('Drug Group:', shapiro(drug_group))
print('Placebo Group:', shapiro(placebo_group))

print("\nLevene's Test for Equality of Variances:")
print(levene(drug_group, placebo_group))

# Step 3: Perform the t-test
t_stat, p_value = ttest_ind(drug_group, placebo_group, equal_var=True)  # Set equal_var=False if variances are unequal
print("\nT-test Results:")
print("t-statistic:", t_stat)
print("p-value:", p_value)
```

```python
plt.figure(figsize=(12, 6))

# Histogram for Drug group
plt.subplot(1, 2, 1)  # 1 row, 2 columns, first subplot
plt.hist(drug_group, bins=20, color='blue', alpha=0.7, edgecolor='black')
plt.title('Histogram of Drug Group (Age)', fontsize=14)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

# Histogram for Placebo group
plt.subplot(1, 2, 2)  # 1 row, 2 columns, second subplot
plt.hist(placebo_group, bins=20, color='green', alpha=0.7, edgecolor='black')
plt.title('Histogram of Placebo Group (Age)', fontsize=14)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

plt.tight_layout()
plt.show()
```

```python
from scipy.stats import mannwhitneyu

# Perform the Mann-Whitney U test
u_stat, p_value = mannwhitneyu(drug_group, placebo_group, alternative='two-sided')

print("Mann-Whitney U Test Results:")
print(f"U-statistic: {u_stat}")
print(f"P-value: {p_value}")
```

This indicates that there is no statistically significant difference between the ages of the Drug group and the Placebo group. Therefore, any observed differences in age between the two groups are likely due to chance, not due to the treatment.

You can confidently conclude that the treatment and control groups have similar age distributions.

Final Conclusion:

Based on the statistical analyses performed, the treatment (drug) is no more harmful in terms of adverse effects compared to the placebo. Additionally, the age distributions of the two groups are similar, meaning that no age-related bias exists in this study. Therefore, the drug appears to be as safe as the placebo within the parameters of this investigation.
