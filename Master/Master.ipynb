# Is biable to study a Master Degree on 2025?

The purpose of this analysis is to evaluate if is rentable to do a master degree with the actual economy.
The bases on this study is to compare the different master that offers the ENEB (Escuela Internacional de Barcelona) with the actual market. The masters offer from this University are: 

- Big Data y Business Intelligence
- Fintech & Blockchain
- Mercados Financieros y Gestión de Activos
- Digital Business
- Revenue Management y Hospitality
- Project Management
- Coaching y Mindfulness
- Logística y Supply Chain Management
- Gestión y Dirección de Equipos
- Dirección de Recursos Humanos y Gestión del Talento
- Marketing Digital y Analytics
- Dirección Comercial y Marketing
- IA Empresarial

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel('Base_Maestría.xlsx')
df
```

```python
df.info()
df.describe()
```

First, we are going to evaluate te number of opportunities per proffesion according to the database:

```python
#Calcule the mean and see what positions are under and what over

df['Mean_P'] = round(df[['P_Computrabajo','P_OCC','P_Glassdoor','P_Indeed']].mean(axis=1))
df_opportunities = df.sort_values('Mean_P',ascending=True)

#Graphic creation

plt.figure(figsize=(12,6))
bars = plt.barh(df_opportunities['Master'],df_opportunities['Mean_P'], color='skyblue')

for bar in bars:
    plt.text(
        bar.get_width(),              
        bar.get_y() + bar.get_height() / 2,  
        f'{bar.get_width():,.0f}',    
        va='center',                  
        ha='left',                    
        fontsize=10                   
    )

plt.title('Opportunities by Proffesion', fontsize=16)
plt.show()
```

On a first look, we can see tha the works related to the people management present a big demand, this is totally related with the condition of the people management since the principel active of each industry is their human force, for this reason, we can see a big demand on:
- Dirección de equipos, recursos humanos, gestipon de equipos and talent.  

It's important to note that since this jobs are to similar, the data can refer to same positions.

After this, we can onte that the most specially profession required is marketing, what can also be related with the big increase with the user on internet and the door to use the web and algorithm to connect with potential customer for a low cost without the necessy of big campains, this is also reflected on the fact that the profession of Marketing appears three times on this list.

Finally the more required techinval profession looks for something not so specialized as Fintech or IA Empresial, but instead lookf for habilities that can be transfer to projects like Logistic and Project Management.

To the end, we can see that Jobs like Blockchain, Fintech or IA Empreserial despita the big impact and that has these areas on the last years, the level of overspecialization is so big that there are so few positions that require this professions.


Once we review the number of opportunities, it's time to review how rentable are these professions on the salary expectations.

```python
#Calcule the mean salary for the professions and create a data frame sorted

df['Mean_S'] = round(df[['S_Computrabajo','S_Jooble','S_Indeed','S_Glassdoor','S_Talent_Mexico','S_Enter_level','S_Experienced_level']].mean(axis=1))
df_salary = df.sort_values('Mean_S',ascending=True)

#Graphic creation 

plt.figure(figsize = (12,6))
bars = plt.barh(df_salary['Master'],df_salary['Mean_S'])
for bar in bars:
    plt.text(
        bar.get_width(),              
        bar.get_y() + bar.get_height() / 2,  
        f'{bar.get_width():,.0f}',    
        va='center',                  
        ha='left',                    
        fontsize=10                   
    )
plt.title('Salary by Profession',fontsize=16)
plt.show()
```

On this graphic, we can see something interesting related with the fact that some positions that were on the top for opportunities are in the botton on the salary expectations, and vice versa. Jobs like Big Data that were on the postion 14 about jobs opportunities is the second on the salary position, and Gestión de Equipos who were on the fiveth positions is now the number 19. We can see that jobs related with the emergin technologies predominate in salary expectations.

First that all, the masters from ENEB are doubles, so we are going to combine this data to give a better reflection.

```python
# Dictionary creation

masters_combinations = {
    "Big Data & Business Intelligence": ["Big_Data", "Business_Intelligence"],
    "Fintech & Blockchain": ["Fintech", "Blockchain"],
    "Mercados Financieros & Gestión de Activos": ["Mercados_Financieros", "Gestión_de_Activos"],
    "Digital Business": ["Digital_Business"],
    "Revenue Management & Hospitality": ["Revenue_Management", "Hospitality"],
    "Project Management": ["Project_Management"],
    "Coaching & Mindfulness": ["Coaching", "Mindfulness"],
    "Logística & Supply Chain Management": ["Logística", "Supply_Chain_Management"],
    "Gestión y Dirección de Equipos": ["Gestión_de_Equipos", "Dirección_de_Equipos"],
    "Dirección de Recursos Humanos & Gestión del Talento": ["Dirección_de_Recursos_Humanos", "Gestión_del_Talento"],
    "Marketing Digital & Analytics": ["Marketing_Digital", "Marketing_Analytics"],
    "Dirección Comercial & Marketing": ["Dirección_Comercial", "Marketing"],
    "IA Empresarial": ["IA_Empresarial"]
}

# Data frame

combined_data = []

for combined, individual in masters_combinations.items():
    filtered_data = df[df['Master'].isin(individual)]
    
    mean_salary = filtered_data['Mean_S'].mean()
    total_opportunities = filtered_data['Mean_P'].sum()
    
    combined_data.append({
        "Master": combined,
        "Mean_Salary": mean_salary,
        "Total_Opportunities": total_opportunities
    })

df_combined = pd.DataFrame(combined_data)
df_combined
```

```python
!pip install adjustText

from adjustText import adjust_text
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))

# Crear el gráfico de dispersión
plt.scatter(df_combined['Total_Opportunities'], df_combined['Mean_Salary'], color='blue', alpha=0.7)

# Crear una lista para las etiquetas
texts = []
for i, txt in enumerate(df_combined['Master']):
    texts.append(
        plt.text(
            df_combined['Total_Opportunities'][i], 
            df_combined['Mean_Salary'][i], 
            txt, 
            fontsize=9
        )
    )

# Ajustar las posiciones de las etiquetas automáticamente
adjust_text(
    texts, 
    force_points=0.3,
    force_text=0.5,
    expand_text=(1.2, 1.2),  # Espacio extra alrededor de las etiquetas
)

# Añadir títulos y etiquetas de los ejes
plt.title('Relación entre Salario Promedio y Oportunidades', fontsize=16)
plt.xlabel('Total de Oportunidades', fontsize=12)
plt.ylabel('Salario Promedio', fontsize=12)
plt.grid()
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```
