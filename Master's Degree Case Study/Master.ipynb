# It's viable to study a Master's Degree on 2025?
### Analyzing the relationship between opportunities and salaries in Mexico



The purpose of this analysis is to evaluate the profitability of pursuing a master’s degree in the current economic climate.

This study focuses on comparing the various master’s programs offered by the **ENEB (Escuela Internacional de Negocios Europea de Barcelona)** with current market trends. The master’s programs analyzed include:
- Big Data and Business Intelligence
- Fintech and Blockchain
- Financial Markets and Asset Management
- Digital Business
- Revenue Management and Hospitality
- Project Management
- Coaching and Mindfulness
- Logistics and Supply Chain Management
- Team Management and Leadership
- Human Resources Management and Talent Development
- Digital Marketing and Analytics
- Commercial and Marketing Management
- Business Artificial Intelligence

The database used for this study was compiled through research conducted on various job platforms, focusing on Mexico-based positions. 

The data was collected on **December 27, 2024**, and includes information on salaries and job opportunities for professionals with qualifications related to each master’s program.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel('Base_Maestría.xlsx')
df
```

As shown in the table, we have a dataset that provides information on the number of vacancies available on each platform and the corresponding mean salary.


To begin the analysis, I will first combine the data from different platforms to calculate an average for the number of opportunities. The reason for using the mean instead of the sum is to avoid duplication, as many companies may post the same job across multiple platforms.

```python
#Calcule the mean and see what positions are under and what over

df['Mean_P'] = round(df[['P_Computrabajo','P_OCC','P_Glassdoor','P_Indeed']].mean(axis=1))
df_opportunities = df.sort_values('Mean_P',ascending=True)

#Graphic creation

plt.figure(figsize=(12,6))
bars = plt.barh(df_opportunities['Master'],df_opportunities['Mean_P'], color='skyblue')

for bar in bars:
    plt.text(
        bar.get_width(),              
        bar.get_y() + bar.get_height() / 2,  
        f'{bar.get_width():,.0f}',    
        va='center',                  
        ha='left',                    
        fontsize=10                   
    )

plt.title('Opportunities by Proffesion', fontsize=16)
plt.show()
```

At first glance, we can observe that jobs related to **people management** show high demand. This is directly linked to the importance of managing human resources, as the main asset of any industry is its workforce. For this reason, we see significant demand for positions such as:  
- **Team Management, Human Resources, Leadership, and Talent Development**.

It is important to note that, since these roles are very similar, the data may refer to overlapping or identical job positions.

Furthermore, we observe that one of the most sought-after professions is in **Marketing**, which can be attributed to the significant increase in internet usage and the opportunity to leverage the web and algorithms to connect with potential customers at a low cost, without requiring large-scale campaigns. This is further emphasized by the fact that Marketing-related positions appear three times in the dataset.

On the other hand, the most in-demand **technical professions** tend to focus on transferable skills, such as **Logistics** and **Project Management**, rather than highly specialized fields like **Fintech** or **Business AI**.

Finally, while jobs in areas like **Blockchain**, **Fintech**, or **Business AI** have gained significant attention due to their impact in recent years, the level of overspecialization in these fields limits the number of positions available. This makes these roles less commonly sought in the job market.

After reviewing the number of opportunities, it is now time to analyze the profitability of these professions based on salary expectations.  

The analysis will follow the same approach: we will create a new column that aggregates the salary data from different sources to calculate an average (mean).

```python
# Calculate the mean salary for the professions and create a data frame sorted
df['Mean_S'] = round(df[['S_Computrabajo','S_Jooble','S_Indeed','S_Glassdoor','S_Talent_Mexico','S_Enter_level','S_Experienced_level']].mean(axis=1))
df_salary = df.sort_values('Mean_S', ascending=True)

# Graphic creation
plt.figure(figsize=(12, 6))
bars = plt.barh(
    df_salary['Master'], 
    df_salary['Mean_S'], 
    color='mediumpurple'
)

for bar in bars:
    plt.text(
        bar.get_width(),              
        bar.get_y() + bar.get_height() / 2,  
        f'{bar.get_width():,.0f}',    
        va='center',                  
        ha='left',                    
        fontsize=10                   
    )
plt.title('Salary by Profession', fontsize=16)
plt.xlabel('Mean Salary', fontsize=12)  # Optional: Add x-axis label
plt.show()
```

In this graphic, we observe an interesting trend: some positions that ranked high in terms of job opportunities are at the bottom in salary expectations, and vice versa. For example, jobs like **Big Data**, which ranked 14th in job opportunities, are now second in salary expectations. Conversely, **Team Management**, which was ranked 5th in job opportunities, has dropped to 19th in salary rankings.  

This demonstrates that jobs related to **emerging technologies** tend to dominate in terms of salary expectations.

### Final Question: Is pursuing a master's degree viable?

Before answering this question, it is essential to highlight that the Master's Degrees offered by ENEB are dual programs. To ensure a more accurate analysis, I will merge the data for the corresponding master's programs.

```python
# Dictionary creation

masters_combinations = {
    "Big Data & Business Intelligence": ["Big_Data", "Business_Intelligence"],
    "Fintech & Blockchain": ["Fintech", "Blockchain"],
    "Mercados Financieros & Gestión de Activos": ["Mercados_Financieros", "Gestión_de_Activos"],
    "Digital Business": ["Digital_Business"],
    "Revenue Management & Hospitality": ["Revenue_Management", "Hospitality"],
    "Project Management": ["Project_Management"],
    "Coaching & Mindfulness": ["Coaching", "Mindfulness"],
    "Logística & Supply Chain Management": ["Logística", "Supply_Chain_Management"],
    "Gestión y Dirección de Equipos": ["Gestión_de_Equipos", "Dirección_de_Equipos"],
    "Dirección de Recursos Humanos & Gestión del Talento": ["Dirección_de_Recursos_Humanos", "Gestión_del_Talento"],
    "Marketing Digital & Analytics": ["Marketing_Digital", "Marketing_Analytics"],
    "Dirección Comercial & Marketing": ["Dirección_Comercial", "Marketing"],
    "IA Empresarial": ["IA_Empresarial"]
}

# Data frame

combined_data = []

for combined, individual in masters_combinations.items():
    filtered_data = df[df['Master'].isin(individual)]
    
    mean_salary = filtered_data['Mean_S'].mean()
    total_opportunities = filtered_data['Mean_P'].sum()
    
    combined_data.append({
        "Master": combined,
        "Mean_Salary": mean_salary,
        "Total_Opportunities": total_opportunities
    })

df_combined = pd.DataFrame(combined_data)
df_combined
```

Now that we have our dataset reflecting the master's degrees offered by ENEB, we can visualize the data to draw meaningful conclusions.  

The most effective way to illustrate the relationship between job opportunities and salary is through a **scatter plot**, as it allows us to observe correlations and identify patterns between these two variables.

```python
from adjustText import adjust_text

# Scatter Plot Creation
plt.figure(figsize=(10, 6))
plt.scatter(df_combined['Total_Opportunities'], df_combined['Mean_Salary'], color='blue', alpha=0.7)

# Labels List
texts = []
for i, txt in enumerate(df_combined['Master']):
    texts.append(
        plt.text(
            df_combined['Total_Opportunities'][i], 
            df_combined['Mean_Salary'][i], 
            txt, 
            fontsize=9
        )
    )
    
adjust_text(
    texts, 
    force_points=0.3,
    force_text=0.5,
    expand_text=(1.2, 1.2),
)

# Format
plt.title('Relation between Salary and Opportunities', fontsize=16)
plt.xlabel('Total Opportunities', fontsize=12)
plt.ylabel('Mean Salary', fontsize=12)
plt.grid()
plt.tight_layout()
plt.show()
```

From this scatter plot, we can conclude that the jobs with the best balance between salary and opportunities are **Digital Business**, **Big Data & Business Intelligence**, and **Project Management**. On the other hand, the roles with the highest number of job opportunities are those related to **human management**.  

As expected, **Coaching & Mindfulness** stands out as the least in-demand and lowest-paying field. While these soft skills can be valuable on a personal level, they do not seem to offer significant value to companies in the current job market.  

### Is pursuing a Master's Degree viable?

The viability of pursuing a master's degree depends on the expected salary of each profession. For instance, if we compare the salaries of these professions to the average salary in Mexico, the result is as follows:

```python
mean_salary = 117996

plt.figure(figsize = (12,6))
bars = plt.bar(df_combined['Master'],df_combined['Mean_Salary'], color = 'cornflowerblue')
plt.xticks(rotation=90)
plt.title('Comparison of salaries against the average salary in Mexico',fontsize = 12)
plt.axhline(y=mean_salary, color='red', linestyle='--', linewidth=2, label=f'Mean: {mean_salary:,}')
for bar in bars:
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        bar.get_height() + 5000,
        f'{bar.get_height():,.0f}',
        ha='center',
        va='bottom',
        fontsize=10
    )

plt.legend()
plt.show()
```

### Final Conclusion

On average, pursuing a Master's Degree appears to be a worthwhile investment for profressionals in Mexico, as it generally leads to higher salary expectations compared to the national average. However, the benefits can vary significantly depending on the specific master's program chosen.  

For example:  
- Programs related to **emerging technologies**, such as **Digital Business** and **Big Data & Business Intelligence**, offer high salary potential but may have fewer job opportunities.  
- On the other hand, programs focused on **human management**, such as **Team Management** and **Human Resources**, provide more opportunities but lower salaries on average.  

It is crucial to carefully evaluate individual goals, industry trends, and job market demands when selecting a master's program to ensure the best return on investment.
